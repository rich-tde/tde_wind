```python
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = [8,6]
plt.rcParams['text.usetex'] = True
plt.rcParams['figure.dpi'] = 200
plt.rcParams.update({'font.size': 16})
gamma = 5/3
noshock = 0.5
mach1 = 1
weak_shock = 1.5
strong_shock = 10
```

```python
""" FUNCTIONS """
def pressure_bump(mach):
    """ P_post/ P_pre shock"""
    Pbump = (2 * gamma * mach**2 - (gamma-1)) / (gamma+1)
    return Pbump

def density_bump(mach):
    """ rho_post/ rho_pre shock"""
    rhobump = (gamma+1) * mach**2 / (mach**2 * (gamma-1) + 2)
    return rhobump

def temperature_bump(mach):
    """ T_post/ T_pre shock"""
    Tbump =  (mach**2 * (gamma-1) + 2) * (2 * gamma * mach**2 - (gamma-1)) / (mach**2 * (gamma+1)**2)
    return Tbump
```

```python
mach_array = np.arange(0.5,10,0.1)
```

```python
""" PLOT """
ax1 = plt.subplot(221)
ax1.plot(mach_array, pressure_bump(mach_array), c = 'green', label = r'$P_1/P_0$')
#ax1.set_xlabel(r'$\mathcal{M}$')
ax1.set_ylabel(r'$P_1/P_0$')

ax2 = plt.subplot(222)
ax2.plot(mach_array, density_bump(mach_array), c = 'b', label = r'$\rho_1/\rho_0$')
#ax2.set_xlabel(r'$\mathcal{M}$')
ax2.set_ylabel(r'$\rho_1/\rho_0$')

ax3 = plt.subplot(212)
ax3.plot(mach_array, temperature_bump(mach_array), c = 'r', label = r'$T_1/T_0$')
ax3.set_xlabel(r'$\mathcal{M}$')
#ax3.set_ylim(0,1)
ax3.set_ylabel(r'$T_1/T_0$')
```
